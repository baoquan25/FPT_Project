```python
import pandas as pd

# Load the dataset
data = pd.read_csv('/content/attitude_data.csv')

# Display the first few rows of the dataset to understand its structure
data.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Create a scatter plot using seaborn
plt.figure(figsize=(8, 8))
sns.scatterplot(x='income_K', y='age', hue='CENTROID_ID', palette='viridis', data=data, s=50)

# Customize the plot
plt.xlabel('Income_K')
plt.ylabel('Age')
plt.title('Clustering Visualization based on Income and Age')

# Add legend title
plt.legend(title='Centroid ID', bbox_to_anchor=(1.05, 1), loc='upper left')

plt.show()
```
