#**Phân tích tốc độ di chuyển trung bình theo khuc vực**
Thống kê tốc độ di chuyển trung bình theo khu vực chỉ xét các cặp quận khởi hành - quận đến có ít nhất 300 chuyến xe trở lên . Vẽ biểu đồ heatmap thể hiện số liệu thống kê trên và rút ra nhận xét


```python
# Lọc dữ liệu để chỉ bao gồm các cặp quận đi - quận đến có ít nhất 300 chuyến
filtered_taxi_data = taxi.groupby(['PUBorough', 'DOBorough']).filter(lambda x: len(x) >= 300)
# nhóm theo quận đi và quận đến, lọc ra những cặp quận đi - đến có >= 300 chuyến

# Tạo crosstab
crosstab = pd.crosstab(filtered_taxi_data['PUBorough'], filtered_taxi_data['DOBorough'], values=filtered_taxi_data['average_speed_mph'], aggfunc='mean')
crosstab
```

```python
plt.figure(figsize=(10, 8))
sns.heatmap(crosstab, cmap = 'coolwarm_r', annot = True, fmt = '.1f')
plt.title('Tốc độ di chuyển trung bình')
plt.show()
```
