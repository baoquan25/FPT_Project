Câu 2: (1.0 điểm)
Cho một danh sách gồm các điểm số của sinh viên trong một lớp học. Hãy tạo một danh sách mới để (-> chứa) xếp loại cho các điểm số tương ứng theo quy tắc sau:
- Điểm < 5:  Yếu
- Điểm >= 5: trung bình
- Điểm >= 6: trung bình khá
- Điểm >= 7: tốt
- Điểm >= 8: giỏi
- Điểm >= 9: xuất sắc
Lưu ý không làm tròn điểm số khi xếp loại.


```python
Diem_so = [4,6,7,3,9,5,8,10,7.5,9.5]
def xep_loai(diem):
    if diem < 5:
        return "Yếu"
    elif diem < 6:
        return "Trung Bình"
    elif diem < 7:
        return "Trung Bình Khá"
    elif diem < 8:
        return "Tốt"
    elif diem < 9:
        return "Giỏi"
    else:
        return "Xuất sắc"
Xep_loai = [xep_loai(diem) for diem in Diem_so]
print(Xep_loai)
```

2. Phân tích đơn biến cho toàn bộ các cột (đếm số giá trị unique, xử lý giá trị null, NA, outliers… nếu có).

```python
for column in age_prediction.columns:
    print(f"Số giá trị unique cột {column}: {age_prediction[column].nunique()}")
```

```python
age_prediction.isnull().sum()
```

```python
def detect_outliers_iqr(df):
    outliers = {}
    for column in df.select_dtypes(include=['int64', 'float64']).columns:
        Q1 = df[column].quantile(0.25)
        Q3 = df[column].quantile(0.75)
        IQR = Q3 - Q1
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR
        outliers[column] = df[(df[column] < lower_bound) | (df[column] > upper_bound)][column]
    return outliers

# Xác định các outliers trong dữ liệu
outliers = detect_outliers_iqr(age_prediction)

# Hiển thị các outliers cho mỗi cột
for column, outlier_values in outliers.items():
    print(f"\nOutliers in column {column}:")
    print(outlier_values)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
age_prediction[['RIDAGEYR','BMXBMI']].corr()
sns.regplot(x='RIDAGEYR',y='BMXBMI',data=age_prediction)

# Thêm tiêu đề cho biểu đồ và các trục
plt.title("Distribution of Age and BMI")
plt.xlabel("Respondent's Age")
plt.ylabel("Respondent's Body Mass Index")

plt.show()
```

Kết luận:
Khi quan sát hình vẽ chúng ta thấy tuổi tác không có mối tương quan với BMI.
Tuy nhiên kết quả tương quan giữa tuổi và chỉ số BMI khi chạy câu lệnh corr, kết quả = 0.147163
cho thấy mối quan hệ giữa hai biến này là rất yếu.


---




```python
def minmaxscaling(x):
  return (x-x.min())/(x.max() - x.min())

age =  minmaxscaling(age_prediction['RIDAGEYR'])
bmi =  minmaxscaling(age_prediction['BMXBMI'])

# thêm nhãn hiệu
plt.title("Distribution of Age and BMI")
plt.xlabel("Value")
plt.ylabel("Frequency")

sns.histplot(age, kde=True, color='lightskyblue', label= 'Age')
sns.histplot(bmi, kde=True, color='darkorange', label= 'BMI')
plt.grid(True, color = 'white')
plt.gca().set_facecolor('#c5eefa')
for spine in plt.gca().spines.values():
    spine.set_visible(False)
plt.gca().set_axisbelow(True)
plt.legend()
plt.show()
```

Kết luận: data có độ tuổi tập trung nhiều nhất trong khoảng từ 10 đến 20 tuổi

6. Phân tích sự khác biệt về chỉ số BMI giữa các nhóm: nhóm có bệnh tiểu đường, nhóm không có bệnh tiểu đường và nhóm chưa chắc chắn, cần xét nghiệm thêm ? Vẽ biểu đồ minh họa và nhận xét

```python
valid_categories = [1, 2, 3]  # 1 = diabetes, 2 = non-diabetes, 3 = not sure
filtered_data = age_prediction[age_prediction['DIQ010'].isin(valid_categories)]

diabetes_bmi = filtered_data[filtered_data['DIQ010'] == 1]['BMXBMI']
non_diabetes_bmi = filtered_data[filtered_data['DIQ010'] == 2]['BMXBMI']
not_sure_bmi = filtered_data[filtered_data['DIQ010'] == 3]['BMXBMI']

data_to_plot = [diabetes_bmi, non_diabetes_bmi, not_sure_bmi]
labels = ['Diabetes', 'Non-Diabetes', 'Not Sure']
colors = ['blue','orange','green']

plt.boxplot(data_to_plot, labels=labels, patch_artist=True, showmeans=True)
plt.title('BMI by Diabetes Status')
plt.xlabel('DIQ010')
plt.ylabel('BMI')
plt.grid(axis='y')
plt.show()
```
